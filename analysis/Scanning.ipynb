## Data testing

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython import display
import networkx as nx
from nxpd import draw
from nxpd import nxpdParams
nxpdParams['show'] = 'ipynb'
```

```python
df = pd.DataFrame.from_csv('../data/scan.csv')
df['signal'] = df['signal'].str.strip(' dBm').astype('float')
df['label'] = df.apply(lambda x: '\n'.join([x['dev'], str(x['signal'])]), axis=1)
df.head()
```

```python
df_twist = df[df['ssid'] == 'twist-test']
df_twist
```

```python
G = nx.from_pandas_dataframe(
    df_twist[df_twist['freq'] == 2412],
    'ap', 'node',
    ['signal', 'label'],
    create_using=nx.DiGraph())
pos = nx.circular_layout(G)
P = nx.nx_pydot.to_pydot(G)
display.display(display.Image(P.create_png()))
```

```python
G = nx.MultiDiGraph()
G.add_nodes_from(list(df_twist['ap'].unique()))
for row in df_twist[df_twist['freq'] == 2412].iterrows():
#     print(row[0])
    G.add_edge(row[1]['ap'], row[1]['node'], row[1]['signal'])
```

```python
P = nx.nx_pydot.to_pydot(G)
display.display(display.Image(P.create_png()))
```

```python
nx.draw_networkx(G, 
    pos,
    node_color='#A0CBE2', 
#     edge_color=range(7),
    with_labels=True,
#     width=4,
#     edge_cmap=plt.cm.Blues
    )
nx.draw_networkx_edge_labels(G, pos)
plt.show()
```
